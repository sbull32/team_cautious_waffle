```python
# Import classes
from sportsipy.nfl.teams import Teams
from sportsipy.nfl.roster import Player
from sportsipy.nfl.roster import Roster

# Function to get player info from Player class object
def get_player_df(player):
    
    # Get player df and add identification
    player_df = player.dataframe
    player_df['player_id'] = player.player_id
    player_df['name'] = player.name
    player_df.set_index('player_id', drop = True, inplace = True)
    return player_df

# Initialize a list of players
players_collected = []
player_count = 0
players2020_df = []

# Iterate through all teams
for team in Teams(year = 2020).dataframes.index:
    print('\n' + team + '\n')
    
    # Iterate through every player on a team roster
    for player_id in Roster(team, year = 2020, slim = True).players.keys():
                
        # Only pull player info if that player hasn't been pulled already
        if player_id not in players_collected:
              
            # Add to count    
            player_count += 1 
            
            player = Player(player_id)
            player_info = get_player_df(player)
             
            # create players2020_df if not initialized
            if not player_count:
                players2020_df = player_info
                player_count = 1    
                
            # else concatenate to players2020_df
            else:
                players2020_df = pd.concat(players2020_df, axis=0)
        
            # Add player to players_collected,
            players_collected.append(player_id)
            print(player.name)

players2020_df.to_csv('nfl_player_stats_by_season.csv')
```
